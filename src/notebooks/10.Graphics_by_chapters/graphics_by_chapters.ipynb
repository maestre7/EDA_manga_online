```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
palette={"MANGA": "#7986cb", "MANHUA": "#8d6e63", "MANHWA": "#81c784"}
labels = ["MANGA", "MANHWA", "MANHUA"]
```

```python
chapters_df = pd.read_csv("../../data/dataset_chapters.csv", index_col="Unnamed: 0")
chapters_df['date_upload'] = pd.to_datetime(chapters_df['date_upload'])
print(len(chapters_df))
```

```python
multy_chapters_df = pd.read_csv("../../data/dataset_multy_chapters.csv", index_col="Unnamed: 0")
multy_chapters_df['date_upload'] = pd.to_datetime(multy_chapters_df['date_upload'])
print(len(multy_chapters_df))
```

```python
fig_histplot = plt.figure(figsize=(10,10))
sns_histplot = sns.histplot(data=chapters_df, x='date_upload', fill=True, hue="type", palette=palette)
sns_histplot.set(title="By Chapters", xlabel=None)
plt.xticks(rotation=-45, fontsize=10);
```

```python
fig = fig_histplot.get_figure()
fig.savefig('../img/by_chapters.png')
```

```python
total_chapters_df = pd.concat([chapters_df, multy_chapters_df])
len(total_chapters_df)
```

```python
fig_histplot2 = plt.figure(figsize=(10,10))
sns_histplot2 = sns.histplot(data=total_chapters_df, x='date_upload', fill=True, hue="type", palette=palette)
sns_histplot2.set(title="By Multy Upload", xlabel=None)
plt.xticks(rotation=-45, fontsize=10);
```

```python
fig = fig_histplot2.get_figure()
fig.savefig('../img/by_multy_upload.png')
```
