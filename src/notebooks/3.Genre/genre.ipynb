```python
import pandas as pd
genre_set = set()
```

Una vez unido los distintos tipos por demografia pasamos a extraer los generos 

```python
dataset_df = pd.read_csv("../../data/dataset_2.csv", index_col='Unnamed: 0')
print(len(dataset_df))
print(dataset_df.columns)
dataset_df
```

```python
genre_list = [str(genre) for genre in dataset_df.genre] 
[genre_set.update(genre.split(", ")) for genre in genre_list]
genre_set.remove('nan')
len(genre_set)
```

```python
# Creamos una columna por cada genero y la cargamos en true o false
for genre in genre_set:
    dataset_df[genre] = dataset_df["genre"].apply(lambda g: True if genre in str(g).split(", ") else False )
```

```python
# Verificamos que las columnas de los generos esten ha false si no tiene generos
dataset_df_df_test = dataset_df[dataset_df["genre"].isnull()]
#dataset_df_df_test
```
